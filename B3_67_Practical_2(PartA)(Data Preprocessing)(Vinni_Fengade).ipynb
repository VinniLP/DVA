### 7.Convert ethnicity to numerical value

```python
copy2 = pd.read_csv("MathScoreTerm1.csv")
```

```python
# copy2['Ethinicity'].replace({'White American': 0,'European American':1,'Hispanic':2, 'African American':3}, inplace = True)
```

### 8.Fill the missing score for a student to the average of the class

```python
copy2 = pd.read_csv("MathScoreTerm1.csv")
copy2["Score"].isnull().describe()
```

```python
copy2["Score"].isnull().index.tolist()
copy2["Score"].fillna(copy2["Score"].mean(),inplace=True)
```
