#### 2. Find the sales data for all Countries- ‘USA’, ‘France’, ‘Australia’, ‘Spain’ , ‘Canada’, etc .
#### Plot Histogram for all the countries and find the country that has maximum sale of Motorcycles and classic cars.

```python
countries_unique = (sales_csv['COUNTRY'].unique())
countries_unique
```

```python
import statistics
from statistics import mean
```

```python
countries_list = []
for i in countries_unique:
    countries_list.append(i)
    print('Country : ', i)
    country_sales = sales_csv[sales_csv['COUNTRY'] == i]['SALES']
    print('Max Sales = ',country_sales.max())
    print('Mean Sales = ', mean(country_sales))
    print('Sales SD = ', statistics.stdev(country_sales))
    print('\n')
```

```python
country_sales = sales_csv.groupby('COUNTRY')['SALES'].sum()
country_sales
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 8))
country_sales.plot(kind='bar')
plt.title('Histogram for all Countries')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```

### 3. Plot sales data for different products using a multiline plot .

```python
sales_product = sales_csv.groupby('PRODUCTLINE')['SALES'].sum()
sales_product
```

```python
import seaborn as sns
```

```python
motorcycleF = sales_csv["PRODUCTLINE"].isin(["Motorcycles"])
classicF = sales_csv["PRODUCTLINE"].isin(["Classic Cars"])
planeF = sales_csv["PRODUCTLINE"].isin(["Planes"])

mc = sales_csv[motorcycleF].groupby('COUNTRY')['SALES'].sum()
cc = sales_csv[classicF].groupby('COUNTRY')['SALES'].sum()
p = sales_csv[planeF].groupby('COUNTRY')['SALES'].sum()


combined_sales = pd.concat([mc, cc, p], axis=1)
combined_sales.columns = ['Motorcycle Sales', 'Classic Car Sales', 'Plane Sales']

sns.set(style="whitegrid")

plt.figure(figsize=(20, 10))

sns.lineplot(data=combined_sales, markers=True)

plt.title('Sales by Country')
plt.xlabel('Country')
plt.ylabel('Sales')

plt.show()
```

#### 4. Find the city that has minimum amount of dealsize under the “Large” category.

#### 7. Use barchart and pie chart to plot the sales of different product lines each year and present the analysis.

```python
sales_csv.info()
```

```python
for i in sales_csv['YEAR_ID'].unique():
    print('FOR', i)
    yearFilter = sales_csv["YEAR_ID"].isin([i])
    yearWiseSales = sales_csv[yearFilter].groupby("PRODUCTLINE")['SALES'].sum()
    plt.figure(figsize=(8, 8))
    plt.pie(yearWiseSales, labels=yearWiseSales.index, autopct='%1.1f%%')
    plt.title('Pie Chart for Product Line Sales')
    plt.figure(figsize=(8, 4))
    yearWiseSales.plot(kind='bar')
    plt.title('Bar Chart for Product Line Sales')
    plt.xlabel('Country')
    plt.ylabel('Total Sales')
    plt.show()
    
    print('\n')
    
```

#### 8. Use appropriate chart to find the product line that has maximum and minimum total sales.

```python
product_wise_sales = sales_csv.groupby('PRODUCTLINE')['SALES'].sum()
product_wise_sales.plot(kind = 'bar')
plt.xlabel('Product Type')
plt.ylabel('Number of products sold (in mil)')
plt.title('Bar Chart for productiwse sales ')

print('Product with max sales : ', product_wise_sales.idxmax())
print('Product with min sales : ', product_wise_sales.idxmin())
```

#### 11. Plot boxplot for total sales values for each year and present the analysis.

```python
bpData = []
for i in sales_csv['YEAR_ID'].unique():
    yearFilter = sales_csv["YEAR_ID"].isin([i])
    yearWiseSales = sales_csv[yearFilter]['SALES']
    bpData.append(yearWiseSales)

plt.figure(figsize=(8, 8))
plt.boxplot(bpData)
plt.xticks([1, 2, 3], ['2003', '2004', '2005'])
plt.title('BoxPlot for sales of each year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
